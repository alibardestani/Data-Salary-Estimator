```python
import pandas as pd
df = pd.read_csv('glassdoor_jobs.csv')
df.head()
```

```python
df.shape
```

```python
# Salary Parsing
```

```python
df = df[df['Salary Estimate'] != '-1']
df
```

```python
salary = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
salary
```

```python
minus_Kd = salary.apply(lambda x : x.replace('K','').replace('$',''))
minus_Kd
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'per hour' in x.lower() else 0)
df
```

```python
df['hourly'].value_counts()
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary:' in x.lower() else 0)
df;
```

```python
df['employer_provided'].value_counts()
```

```python
min_hr = minus_Kd.apply(lambda x : x.lower().replace('per hour', '').replace('employer provided salary:',''))
min_hr
```

```python
df['min_salary'] = min_hr.apply(lambda x : int(x.split('-')[0]))
df.head()
```

```python
df['min_salary'].dtype
```

```python
df['max_salary'] = min_hr.apply(lambda x : int(x.split('-')[1]))
```

```python
df.info()
```

```python
df['avg_salary'] = (df.min_salary + df.max_salary)/2
df
```

```python
import numpy as np
df.replace('-1', np.nan, inplace=True)
```

```python
df.replace(-1, np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
# Company Name Text Only

df['Company_Txt'] = df.apply(lambda x : x['Company Name'] if x['Rating'] == np.nan else x['Company Name'][:-3], axis=1)
df
```

```python
# State Field

df['Job State'] = df['Location'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
df['Job State'].value_counts()
```

```python
df['Same State'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
# Age Of Company

df['Age'] = df.Founded.apply(lambda x : x if x == np.nan else 2024 - x)
df.head()
```

```python
df.Age.value_counts()
```

```python
# Parsing of Job Description (python, etc.)

# Python 
df['Python_yn'] = df['Job Description'].apply(lambda x : 1 if 'python' in x.lower() else 0)

# R Studio
df['R_yn'] = df['Job Description'].apply(lambda x : 1 if 'r studio' in x.lower() or 'r_studio' in x.lower() else 0)

# Spark
df['Spark_yn'] = df['Job Description'].apply(lambda x : 1 if 'spark' in x.lower() else 0)

# Aws
df['Aws_yn'] = df['Job Description'].apply(lambda x : 1 if 'aws' in x.lower() else 0)

# Excel
df['Excel_yn'] = df['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)

df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.to_csv('salary_data_cleaned.csv', index = False)
```
